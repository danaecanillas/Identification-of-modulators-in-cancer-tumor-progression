```python
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA

path_file = "data/pathwayscores_clinicalinformation_METABRIC.tsv"
data = pd.read_csv(path_file, sep='\t')
```

```python
features = ['Cell_Cycle', 'HIPPO', 'MYC', 'NOTCH', 'NRF2', 'PI3K', 'TGF.Beta', 'RTK_RAS', 'TP53', 'WNT', 'Hypoxia', 'SRC', 'ESR1', 'ERBB2']
target = ['PAM50']
```

```python
# Separating out the features
x = data.loc[:, features].values

# Separating out the target
y = data.loc[:,target].values
```

```python
dim_pca=2
pca = PCA(n_components=dim_pca)

principalComponents = pd.DataFrame(pca.fit_transform(x))

finalDf = pd.concat([data.loc[:,target], principalComponents], axis = 1)
```

```python
# Plot
fig = px.scatter(x = finalDf.iloc[:,1], y = finalDf.iloc[:,2], hover_data=[finalDf.iloc[:,0]], color = finalDf.iloc[:,0], marginal_x = 'box')
fig.show() 
```
