```python
path_file = "data/METABRIC_table_S23_revised.txt"
newdata = pd.read_csv(path_file, sep='\t')
```

```python
newdata =  newdata.rename(columns={'METABRIC_ID': 'submitter'})
newdata = newdata.sort_values(by=['submitter'])
newdata
```

```python
path_file = "data/pathwayscores_clinicalinformation_METABRIC.tsv"
data = pd.read_csv(path_file, sep='\t')
data = data.sort_values(by=['submitter'])
data
```

```python
entire_data = pd.merge(newdata, data, on=['submitter','PAM50','age_at_diagnosis','stage','RFSE','RFS'], how='inner')
entire_data
```

```python
entire_data[["AGE","DSSE10", "DSS10", "DSS10Y", "grade","GradeG", "size", "TsizeG", "NPI","A","B","WNT"]]
```

```python
cor = entire_data.corr()
relevant_features = cor[(abs(cor)>0.4)]
relevant_features

relevant_features.to_csv("data/relevant_features.csv",index=True)
```

```python
cor = entire_data[['RFSE', 'DSSE', 'DSSE10', 'DSS10', 'DSS10Y','grade','TsizeG','Node',
       'lymph_nodes_positive', 'NPI','HER2_IHC_status',
       'HER2', 'PAM50L','RFS', 'A', 'B',
       'LOLO', 'Basal', 'Her2', 'LumA', 'LumB', 'Normal','RORS',
       'PROLIF', 'RORP', 'RORC', 'RORPC']].corr()
mask = np.triu(np.ones_like(cor, dtype=bool))
cmap = sns.color_palette("Spectral_r", as_cmap=True)
sns.set(rc={'figure.figsize':(15.7,15.7)})
#sns.heatmap(cor, cmap=cmap, vmax=1, vmin=-1, mask=mask, center=0, square=True)
sns.clustermap(cor, cmap=cmap, vmax=1, vmin=-1, metric="correlation", dendrogram_ratio=0.05, cbar_pos=(-0.1, 0.8, 0.05, 0.18))
```
